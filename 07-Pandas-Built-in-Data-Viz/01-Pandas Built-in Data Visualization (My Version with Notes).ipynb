## Style Sheets

Matplotlib has [style sheets](http://matplotlib.org/gallery.html#style_sheets) you can use to make your plots look a little nicer. These style sheets include plot_bmh,plot_fivethirtyeight,plot_ggplot and more. They basically create a set of style rules that your plots follow. I recommend using them, they make all your plots have the same look and feel more professional. You can even create your own if you want your company's plots to all have the same look (it is a bit tedious to create on though).

Here is how to use them.

**Before plt.style.use() your plots look like this:**

```python
df1['A'].hist(bins=30) #you can see in the object type, that matplotlib is working under the hood
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

Now your plots look like this:

```python
df1['A'].hist(bins=30)
```

```python
#Another syntax of plotting data directly from dataframe:
#you have to specify the kind of plot you want with the dataframe
df1['A'].plot(kind='hist',bins=30)
```

```python
#Another syntax of plotting data directly from dataframe:
#dataframe['column'].plot.plot_type()
df1['A'].plot.hist()
```

# Plot Types

There are several plot types built-in to pandas, most of them statistical plots by nature:

* df.plot.area     
* df.plot.barh     
* df.plot.density  
* df.plot.hist     
* df.plot.line     
* df.plot.scatter
* df.plot.bar      
* df.plot.box      
* df.plot.hexbin   
* df.plot.kde      
* df.plot.pie

You can also just call df.plot(kind='hist') or replace that kind argument with any of the key terms shown in the list above (e.g. 'box','barh', etc..)

## Histograms

```python
#Most common used plot for pandas dataframes is Histogram
df1['A'].plot.hist(bins=50)
```

## Line Plots

```python
#Another most common plot used for pandas dataframe is line plot
#All the parameters of matplotlib are also applicable here
#Changin figure size.
#Changing line width to make it a bit thinner
df1.plot.line(y='B',figsize=(12,3),lw=1) 
#By default argument x=df1.index(which is the index column of df1)
```

## Scatter Plots

```python
#Scatter plot
#You can add another argument based on color. add c as an argument
df1.plot.scatter(x='A',y='B',c='C') #Essentially it is showing 3D information of 3 columns here
```

```python
#We can also use dot size instead of color to represent the information of the 3rd column
#Points will be too small so we multiply it by some sort of factor say 30
df1.plot.scatter(x='A',y='B',s=df1['C']*30)
```

## BoxPlots

```python
#Box Plots
#Creates box plots for all the columns
df2.plot.box()
```

## Hexagonal Bin Plot

Useful for Bivariate Data, alternative to scatterplot:

```python
df = pd.DataFrame(np.random.randn(1000,2),columns=['a','b'])
```

```python
df.head()
```

```python
# Hexbin plot. 
'''
A Hexbin plot is useful to represent the relationship of 2 numerical variables,
when you have a lot of data point. 
Instead of overlapping, the plotting window is split in several hexbins(Hexagons),
and the number of points per hexbin is counted. The color denotes this number of points.
'''  
df.plot.hexbin(x='a',y='b',gridsize=25,figsize=(10,7))
```

## Kernel Density Estimation plot (KDE)

```python
#KDE plot on a column
df2['a'].plot.kde()
```

```python
#KDE plot on all the columns
df2.plot.kde()
```
