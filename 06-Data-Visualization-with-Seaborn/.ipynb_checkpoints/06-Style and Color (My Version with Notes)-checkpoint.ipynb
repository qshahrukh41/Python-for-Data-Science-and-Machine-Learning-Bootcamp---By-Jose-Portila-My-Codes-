# Style and Color

We've shown a few times how to control figure aesthetics in seaborn, but let's now go over it formally:

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
```

## Styles

You can set particular styles:

```python
#To change style sns.set_style()
#sns.set_style('white') #White Background on back
#sns.set_style('dark') #Dark Background on back
#sns.set_style('whitegrid') #White Background with grid lines
#sns.set_style('ticks') #Ticks on the axes
sns.countplot(x='sex',data=tips)
#sns.despine() #Removes upper and right border/spine around the plot.
#sns.despine(left=True,bottom=True) 
#passing on the parameters(left=True,bottom=True) will remove all for spines/borders around the plot
```

## Size and Aspect

You can use matplotlib's **plt.figure(figsize=(width,height) ** to change the size of most seaborn plots.

You can control the size and aspect ratio of most seaborn grid plots by passing in parameters: size, and aspect. For example:

```python
#Changing size of the plots. Here were are using matplotlib figure sizing to seaborn
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=tips)
```

## Scale and Context

The set_context() allows you to override default parameters:

```python
#Scaling figure size according to context
'''
If we want to create plots to put it on a poster we might want larger figure. 
To do so, sns.set_context('poster')
'''
#If you are still not satisfied with font size, you can scale that as well using font_scale
#By default context is set to 'notebook'
sns.set_context('notebook',font_scale=1)
sns.countplot(x='sex',data=tips)
```

## Palettes and colors

```python
#Palettes and colors
#palette parameters: coolwarm, magma,...etc
#How to find colormap of matplotlib: Goto: https://matplotlib.org/tutorials/colors/colormaps.html
#There you can find names of many colormaps
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='Spectral')
```
