# Regression Plots

Seaborn has many built-in capabilities for regression plots, however we won't really discuss regression until the machine learning section of the course, so we will only cover the **lmplot()** function for now.

**lmplot** allows you to display linear models, but it also conveniently allows you to split up those plots based off of features, as well as coloring the hue based off of features.

Let's explore how this works:

```python
#We'll see linear model plot, lmplot that allows to display linear models with seaborn
#We'll also learn how to size and change the aspect ratio for seaborn plots
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

## lmplot()

```python
#lmplot to create a very simple linear model plot
#Running sns.lmplot will create a scatter plot with a linear fit to the model
'''
Adding a hue will essentially create two scatter plots and 
two linear fits to the data based on a categorical column. For instance: hue='sex'
'''
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm')
```

```python
#You can change the marker types by passing a list of markers. For instance, here markers=['o','v']
'''
You can also change marker size. 
and that's just using the matplotlib function on top of seaborn. 
To do it, we'll pass scatter_kws={'s':100}
'''
#Generally we'll not do so much customization of our plots with seaborn such as to change markersize
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],scatter_kws={'s':100})
```

## Using a Grid

We can add more variable separation through columns and rows with the use of a grid. Just indicate this with the col or row arguments:

```python
'''
Instead of seperating by hue/color,if you want to seperate male female data by grids, 
we simply add col='sex' instead of hue='sex'. That will call grid plots(Facetgrid)
'''
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
'''
You can also add row as another categorical column (ie:row='time') and 
that will essentially create a facetgrid
'''
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time')
```

```python
#Here we call the lmplot but with hue='sex' and col='day'
#We can see here the size of the plots are somewhat unreadable
#To adjust we can change aspect and height parameters
'''
Still the legend names and col names on top are somewhat unreadable. 
We'll learn to adjust them in the next lecture: style and color
'''
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',col='day',aspect=0.8,height=8)
```
