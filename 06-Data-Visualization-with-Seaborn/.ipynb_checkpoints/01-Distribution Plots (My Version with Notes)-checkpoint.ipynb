# Distribution Plots

Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:

* distplot
* jointplot
* pairplot
* rugplot
* kdeplot

```python
import seaborn as sns
```

```python
%matplotlib inline
```

## Data
Seaborn comes with built-in data sets!

```python
'''
seaborn has some built in datasets that you can directly load. 
We're gonna grab one called tips and name it as tips
'''
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## distplot

The distplot shows the distribution of a univariate set of observations.

```python
# dist plot lets us show distribution of a univarite set of observations
# univariate is just the way of saying one variable
sns.distplot(tips['total_bill'],kde=False,bins=30)
# Plotting it we get a histogram and a line showing kernel density estimation
# You can remove kde for now by adding kde=False
# Most of the bills are betn 10 and 20 dollar and fades away as we go higher and higher
# You can add bins to histogram. Lets choose bins=30
# Choosing bins=100, you'll essentially start to see each data as a bin(Not recommeded). 
# We want to balance the binsize
```

## jointplot

jointplot() allows you to basically match up two distplots for bivariate data. With your choice of what **kind** parameter to compare with: 
* “scatter” 
* “reg” 
* “resid” 
* “kde” 
* “hex”

```python
# Joint plot. It allows you to basically plot bivariate data. 
# Means it allows to join two different distribution plots
# Lets say, we want to compare distribution of total bill vs tips size
sns.jointplot(x='total_bill',y='tip',data=tips)
# We can see a trend that as we go higher on total bill, we go higer on tip
# kind' parameter  allows us to see what's actually going on inside of the joint plot
# As default 'kind' is set to scatter. But we can also set it as kind='hex'
'''
kind='hex' lets us see a hexagon distribution representation. 
It's similar to scatter plot; except if hexagon has a certain number of points it gets darker. 
and if number of points is less it gets lighter
'''
# kind='reg' will plot a regression line along with scatter plot
# kind='kde' will allow to show the density where these points match up most
# Default kind scatter is usually the best way to look at data.
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

## pairplot

pairplot will plot pairwise relationships across an entire dataframe (for the numerical columns) and supports a color hue argument (for categorical columns). 

```python
'''
Pairplot is essentially used to plot pairwise relationships,
accross an entire dataframe at least for the numerical columns
'''
#It also supports a color hue argument for categorical columns which we'll see later on.
'''
Pairplot essentially does what jointplots does,
but for every single possible numerical pair of columns in the dataset
'''
#Pairplot is something we're going to do quite a bit throughout the course
#The larger the dataframe the longer pairplots take to execute
'''
#In the pairplot when a column is plotted against itself, 
instead of a scatter plot we basically see a histogram. 
That is the right diagonal elements will be histograms.
'''
'''
You can pass categorical attributes as hue. 
for instance hue='sex' will describe the plots in the same way,
but for each male and female we'll see a different color
'''
'''
As a third argument you can pass in palette. 
Palette allows to color the plots with some specific color. 
It is going to choose some specific colors for the plots

'''
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

## rugplot

rugplots are actually a very simple concept, they just draw a dash mark for every point on a univariate distribution. They are the building block of a KDE plot:

```python
#rugplots is to actually explain what the kde plots where.
'''
just like distribution plots we're going to pass in a single column here. 
For instance sns.rugplot(tips['total_bill'])
'''
'''
rug plot is very simple. 
For every data for a specific column it just puts a dash '|' in the plot 
'''
sns.rugplot(tips['total_bill'])
```

## Again Distplot

```python
'''
# Comparing rugplot with histogram: 
histogram has bins and it counts how many dashes were in that bin and
then shows it as a number in y column.
'''
# Those regions where there is high density of dashes, for those histogram bins will be taller
# We want to build off by the idea of rug plots to actually explain what kde is
# KDE stands for kernel density estimation plots.
'''
For each of the dashes(data) in rugplot,
we take a normal distribution curve around it taking it as the center and
summing all those curves will finally give us the KDE plot.
'''
# Let's see how KDE plot is generated by seeing an detailed example in the next cell.
sns.distplot(tips['total_bill'])
```

### KDE Plot

```python
# Don't worry about understanding this code!
# It's just for the diagram below
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

#Create dataset
dataset = np.random.randn(25)

# Create another rugplot
sns.rugplot(dataset);

# Set up the x-axis for the plot
x_min = dataset.min() - 2
x_max = dataset.max() + 2

# 100 equally spaced points from x_min to x_max
x_axis = np.linspace(x_min,x_max,100)

# Set up the bandwidth, for info on this:
url = 'http://en.wikipedia.org/wiki/Kernel_density_estimation#Practical_estimation_of_the_bandwidth'

bandwidth = ((4*dataset.std()**5)/(3*len(dataset)))**.2


# Create an empty kernel list
kernel_list = []

# Plot each basis function
for data_point in dataset:
    
    # Create a kernel for each point and append to list
    kernel = stats.norm(data_point,bandwidth).pdf(x_axis)
    kernel_list.append(kernel)
    
    #Scale for plotting
    kernel = kernel / kernel.max()
    kernel = kernel * .4
    plt.plot(x_axis,kernel,color = 'grey',alpha=0.5)

plt.ylim(0,1)
```

```python
# To get the kde plot we can sum these basis functions.

# Plot the sum of the basis function
sum_of_kde = np.sum(kernel_list,axis=0)

# Plot figure
fig = plt.plot(x_axis,sum_of_kde,color='indianred')

# Add the initial rugplot
sns.rugplot(dataset,c = 'indianred')

# Get rid of y-tick marks
plt.yticks([])

# Set title
plt.suptitle("Sum of the Basis Functions")
```
