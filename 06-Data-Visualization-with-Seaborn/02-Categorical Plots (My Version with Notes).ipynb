# Categorical Data Plots

Now let's discuss using seaborn to plot categorical data! There are a few main plot types for this:

* factorplot
* boxplot
* violinplot
* stripplot
* swarmplot
* barplot
* countplot

Let's go through examples of each!

```python
#Plotting categorical data
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
import numpy as np
```

## barplot and countplot

These very similar plots allow you to get aggregate data off a categorical feature in your data. **barplot** is a general plot that allows you to aggregate the categorical data based off some function, by default the mean:

```python
#Barplot (Most basic categorical plot)
'''
Barplot is a general plot that aggregate the categorical data based off some function
and by default that function is mean()
'''
#You can think of it as visualization by group by action
'''
If we choose categorical column x='sex' and y='total_bill' for barplot,
then the barplot will show us mean value of total bill based on gender.
'''
#We see that average bill for male is slightly higher than female
'''
We can change the aggregate function from mean to -
something else by adding another argument estimator.
'''
'''
Let we want to see the standard deviation of total_bill based on gender. 
To see if there is more variation of bills in male or female
'''
#For standard deviation as estimator we need to import numpy
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
#Let'see another categorical plot namely countplot
#Explicitly counts number of occurances of each category for a categorical column
#Thus we only pass one parameter which is category
#We see there are more male than female in the dataset
sns.countplot(x='sex',data=tips)
```

```python
#Boxplot. Also known as box and whisker plot
#The points which are out of the whisker are known as outliers
# x-> categorical and y->numerical data
#lets draw a barplot for x='day' and y='total_bill'
'''
#We can add another categorical value. Let, hue='smoker'. 
Then the boxplot will show box and whisker plots of total bill,
with respect to if the person is a smoker or not.
'''
'''
We can derive some information from this. 
Such as people in Friday tend to have more bill if they are not a smoker than if they are a smoker
'''
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
#Violin plot. It plays a very similar role as boxplot
#It takes exact same parameters as boxplot. 
# x-> categorical and y->numerical data
# Violinplot allows us to actually plot all the components that correspond to actual datapoints
# Basically it shows the kernel density estimation of the underlying distribution
'''
It allows to reveal more information about the data points itself, 
as it reveals information of all the data points
'''
# But it's just a little harder to interpret
# Violinplot also has a option to do hue
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
#You can directly compare the hue values by using another argument split=True
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker',split=True)
```

```python
#You essentially have to know your audience to use these advanced visualizations.
#Because if they don't understand what you are trying to deliver it's of no use
'''
For simply showing it to CEO or management, 
a simple boxplot would suffice rather than using violinplots
'''
```

```python
#stripplot
#stripplot draws a scatterplot where one variable is categorical

sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
'''
You can also add hue in stripplots and then split(now named as dodge),
to see the data points more clearly
'''
#Stripplot can be a bit harder to interpret than boxplot or violinplot
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',dodge=True)
```

```python
#Some people like the idea of combining violinplot vs stripplot and that's known as swarmplot
sns.swarmplot(x='day',y='total_bill',data=tips)
'''
#For very large datasets its very hard to arrange the data points so,
they don't stack on top of one another. Thus computationally expensive.

'''
#Overall, mostly stay away from swarmplots, since most people don't know much of them
```

```python
#You can also stack swarm plots on top of violinplots. But computationally expensive
'''
Be careful it is not so useful for presentation, 
but sometimes just for exploratory data analysis
'''
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
'''
Basically all the plots we have learned we can generally use any of them using catplot 
and specifying kind='plottype'(such as:violin,bar, etc) as an argument.
'''
sns.catplot(x='day',y='total_bill',data=tips,kind='violin')
sns.catplot(x='day',y='total_bill',data=tips,kind='bar')
```
