# Grids

Grids are general types of plots that allow you to map plot types to rows and columns of a grid, this helps you create similar plots separated by features.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

## PairGrid

Pairgrid is a subplot grid for plotting pairwise relationships in a dataset.

```python
#using PairGrid we can gain a lot more control over the pairplots
#Just calling PairGrid will give us blank axes for all the numerical feature pairs
#pairplot is the simplified version of PairGrid
'''
pairplot does lot of the stuff automatically, 
in PairGrid you have to a little more but it gives you a lot more control
'''
#Store the pairgrid in a variable. let, g=sns.PairGrid(iris)
#Then with that variable you can map plot type. For instance, g.map(plt.scatter)
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
#You can also specify what you want in diagonals, on the upper half and on the lower half
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
#If you want customized pairplots for your dataset, then you should go for PairGrids
```

## pairplot

pairplot is a simpler version of PairGrid (you'll use quite often)

```python
sns.pairplot(iris) #Pairplot is the automated jointplot for all the columns in the dataset
```

## Facet Grid

FacetGrid is the general way to create grids of plots based off of a feature:

```python
#facet grid
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Just the Grid
g = sns.FacetGrid(tips, col="time", row="smoker")
```

```python
'''
Specify the data for FacetGrid. Specify two features. 
one as column and one as row. Here, col='time' and row='smoker'
'''
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')
```

```python
'''
You can also use facetgrid for plots with two vaariables. 
For instace, the previous distplot was only for total_bill.
'''
#Now we'll use scatter plot with 'total_bill' and 'tip' using the previous facetgrid
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(plt.scatter,'total_bill','tip')
```
