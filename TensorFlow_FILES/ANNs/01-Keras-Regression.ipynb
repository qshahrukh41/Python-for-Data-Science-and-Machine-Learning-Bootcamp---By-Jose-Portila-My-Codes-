#### Feature Columns
    
* id - Unique ID for each home sold
* date - Date of the home sale
* price - Price of each home sold
* bedrooms - Number of bedrooms
* bathrooms - Number of bathrooms, where .5 accounts for a room with a toilet but no shower
* sqft_living - Square footage of the apartments interior living space
* sqft_lot - Square footage of the land space
* floors - Number of floors
* waterfront - A dummy variable for whether the apartment was overlooking the waterfront or not
* view - An index from 0 to 4 of how good the view of the property was
* condition - An index from 1 to 5 on the condition of the apartment,
* grade - An index from 1 to 13, where 1-3 falls short of building construction and design, 7 has an average level of construction and design, and 11-13 have a high quality level of construction and design.
* sqft_above - The square footage of the interior housing space that is above ground level
* sqft_basement - The square footage of the interior housing space that is below ground level
* yr_built - The year the house was initially built
* yr_renovated - The year of the houseâ€™s last renovation
* zipcode - What zipcode area the house is in
* lat - Lattitude
* long - Longitude
* sqft_living15 - The square footage of interior housing living space for the nearest 15 neighbors
* sqft_lot15 - The square footage of the land lots of the nearest 15 neighbors

## Scaling and Train Test Split

```python
X = df.drop('price',axis=1)
y = df['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

### Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
#Difference between fit_transform and transform:
#To center the data (make it have zero mean and unit standard error), 
#you subtract the mean and then divide the result by the standard deviation.
#You do that on the training set of data. But then you have to apply the same transformation to your testing set (e.g. in cross-validation), or to newly obtained examples before forecast. 
#But you have to use the same two parameters Î¼ and Ïƒ (values) that you used for centering the training set.
#Hence, every sklearn's transform's fit() just calculates the parameters (e.g. Î¼ and Ïƒ in case of StandardScaler) and 
#saves them as an internal objects state. 
#Afterwards, you can call its transform() method to apply the transformation to a particular set of examples.
```

```python
X_train= scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train.shape
```

```python
X_test.shape
```

## Creating a Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import Adam
```

```python
#We have 19 features in our data and it's essentially a good choice of nodes in each layer
#Later on we'll see how to choose the number of layers in the deep learning model
model = Sequential()

model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')
```

## Training the Model

```python
#Here we added a parameter validation_data using our test set to see how well the model is performing not only on training set, but also on test/validation data
#Note that testing performance on validation_data will not add biases to our model. It just checks how the model is doing on test data
#y_train and y_test were previously not converted to numpy array, which we have done here. Because tensorflow essentially works with numpy
#Using batch size=128, we do not train the model using our entire training data. Instead we train in batches. It prevents overfitting the training data
model.fit(x=X_train,y=y_train.values,
          validation_data=(X_test,y_test.values),
          batch_size=128,epochs=400)
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
#We can see both training loss and validation loss is decreasing
#Also, there have been no increase so far in the validation set. That indicates we could continue training
#If somehow validation loss starts going up, that's when we know that we are overfitting and we need to stop
losses.plot()
```
